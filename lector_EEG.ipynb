```python
import pyxdf
# Replace 'your_file.xdf' with the path to your XDF file
file_path = '/mnt/d/Florencia/Desktop/Facultad/Tesis/TesisFlor/Mediciones/S02_auditivo.xdf'
#Archivos computadora Flor
#'/mnt/d/Florencia/Downloads/S01_auditivo.xdf'
#'/mnt/d/Florencia/Downloads/S01_visual.xdf'

# Load the XDF file
streams, header = pyxdf.load_xdf(file_path)

# Iterate through streams to find the marker stream
for stream in streams:
    if stream['info']['type'][0] == 'Markers':
        marker_stream = stream
        break
else:
    raise ValueError("No marker stream found in the XDF file.")

# Access marker information
markers = marker_stream['time_series']
marker_timestamps = marker_stream['time_stamps']

# Print the markers and their timestamps
for timestamp, marker in zip(marker_timestamps, markers):
    print(f'Timestamp: {timestamp}, Marker: {marker}')
```

```python
from PIL import Image

def convert_tif_to_jpg(input_path, output_path):
    try:
        # Open the TIFF file
        with Image.open(input_path) as img:
            # Save the image in JPEG format
            img.convert('RGB').save(output_path, 'JPEG')
            print(f"Conversion successful. Saved as {output_path}")
    except Exception as e:
        print(f"Error: {e}")

# Specify the input and output file paths
input_file_path = '/mnt/d/Florencia/Desktop/Trabajo/ESTUDIOS RIESGO HIDRICO/CN II/CN II ap2.tif'
output_file_path = '/mnt/d/Florencia/Desktop/Trabajo/ESTUDIOS RIESGO HIDRICO/CN II/CN II ap2.jpg'

# Call the function to convert the TIFF to JPEG
convert_tif_to_jpg(input_file_path, output_file_path)
```
