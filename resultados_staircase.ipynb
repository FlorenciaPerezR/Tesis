```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
```

```python
sujnum = ['02','03','04','05','06','07','08','09','10','11','12']
volumen_total = []
contraste_total = []
for num in sujnum:
    star_audi = genfromtxt('/home/nicobruno/flor/Tesis/staircase/S'+num+'_staircase_auditivo.csv', skip_header = 1, delimiter=';')
    star_vis = genfromtxt('/home/nicobruno/flor/Tesis/staircase/S'+num+'_staircase_visual.csv', skip_header = 1, delimiter=';')
    volumen_total.append(star_audi[:,1])
    contraste_total.append(star_vis[:,3])
    volumen = star_audi[:,1]
    trial_audi = star_audi[:,0]
    contraste = star_vis[:,3]
    trial_vis = star_vis[:,0]
    plt.figure(1)
    grafico_auditivo= plt.plot(trial_audi, volumen, label = num)
    plt.title('Audity Staircase')
    plt.ylabel('Volume')
    plt.xlabel('Trial')
    plt.legend()

    #plt.figure(2)
    #grafico_visual = plt.plot(trial_vis, contraste, label = num)
    #plt.title('Visual Staircase')
    #plt.ylabel('Contrast')
    #plt.xlabel('Trial')
    #plt.legend()

# Guardar los gráficos
plt.savefig('/home/nicobruno/flor/Tesis/graficos/staircase_auditivo_ingles.png')

#plt.savefig('/home/nicobruno/flor/Tesis/graficos/staircase_visual_ingles.png')
```

```python
# Calcular el rate (pendiente) entre puntos consecutivos
rate_auditivo = np.diff(volumen) / np.diff(trial_audi)

# Graficar el rate
plt.figure()
plt.plot(trial_audi, volumen, label = num)  # trial_audi[:-1] para alinear con las diferencias
plt.plot(trial_audi,rate_auditivo)
plt.title('Rate de cambio - Auditivo')
plt.ylabel('Rate (ΔVolumen / ΔTrial)')
plt.xlabel('Trial')
plt.show()
```

```python
rate_visual = np.diff(contraste) / np.diff(trial_vis)

plt.figure()
plt.plot(trial_vis[:-1], rate_visual)  # trial_vis[:-1] para alinear con las diferencias
plt.title('Rate de cambio - Visual')
plt.ylabel('Rate (ΔContraste / ΔTrial)')
plt.xlabel('Trial')
plt.show()
```
