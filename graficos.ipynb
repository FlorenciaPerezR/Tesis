```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
```

```python
auc = '/home/nicobruno/flor/Tesis/modelo_resultados/all_auc_iter_1000_e_100_visual.csv'
auc_sh = '/home/nicobruno/flor/Tesis/modelo_resultados/all_auc_sh_iter_1000_e_100_visual.csv'

auc_datos= genfromtxt(auc, skip_header=1,delimiter=',')
auc_sh_datos= genfromtxt(auc_sh, skip_header=1,delimiter=',')
```

```python
path = '/home/nicobruno/flor/Tesis/graficos/'
data = [auc_datos, auc_sh_datos]

fig = plt.figure(figsize =(10, 7))

# Creating plot
plt.boxplot(data)
plt.xticks([1, 2], ['auc', 'auc_sh'])
# Adding title 
plt.title("Comparaci√≥n AUC y AUC_SH caso visual para 1000 iter")
# show plot
plt.show()

fig.savefig('/home/nicobruno/flor/Tesis/graficos/auc_visual_1000.png')
#/home/nicobruno/flor/Tesis/graficos 
#/home/nicobruno/flor/Tesis/graficos 
```
