```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from numpy import genfromtxt

auc = '/home/nicobruno/flor/Tesis/modelo_resultados/all_auc_iter_100_e_100_auditivo.csv'
auc_sh = '/home/nicobruno/flor/Tesis/modelo_resultados/all_auc_sh_iter_100_e_100_auditivo.csv'

auc_datos= genfromtxt(auc, skip_header=1,delimiter=',')
auc_sh_datos= genfromtxt(auc_sh, skip_header=1,delimiter=',')

# Suponiendo que tienes listas con los valores de AUC
auc_with_shuffling = auc_sh_datos
auc_without_shuffling = auc_datos

# Crear el histograma
plt.figure(figsize=(8, 6))
sns.histplot(auc_with_shuffling, bins=30, color='red', label='Con permutación', alpha=0.6)
sns.histplot(auc_without_shuffling, bins=30, color='blue', label='Sin permutación', alpha=0.6)

# Agregar etiquetas y leyenda
plt.xlabel('AUC')
plt.ylabel('Frecuencia')
plt.title('Histogramas de valores de AUC con y sin permutación de etiquetas')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
#Hago el grafico de los features pero embellecido

import numpy as np
from numpy import genfromtxt

dir = '/home/nicobruno/flor/Tesis/modelo_resultados/'
pvals = genfromtxt(dir+'nuevo_pvals_iter_100_e_100_auditivo.csv', skip_header=1,delimiter=',')

# Convertir p-vals en array de NumPy
pvals = np.array(pvals)

features = ['wSMI_1','wSMI_2','wSMI_4','wSMI_8','p_e_1','p_e_2','p_e_4','p_e_8','k',
            'se','msf','sef90','sef95','b','b_n','g','g_n','ft','ft_n','t',
            't_n','d','d_n','a_n','a']

fi_df = pd.read_csv(dir+'/nuevo_fi_iter_100_e_100_auditivo.csv')
fi_df = fi_df.melt(var_name="feature", value_name="value")

pvals_df = pd.DataFrame({"feature": features, "pval": pvals})
df = pvals_df.merge(fi_df, on="feature")
df["grupo"] = np.where(df["pval"] < 0.05, "significant", "non_significant")

palette = ['gray', 'g']

fig, ax = plt.subplots(figsize=(8, 6))


# Boxplot con transparencia en el relleno
sns.boxplot(data=df, x='feature', y='value', hue='grupo', palette=palette,
            width=0.7, dodge=False, ax=ax, zorder=1, boxprops=dict(alpha=.5))

# Puntos individuales (stripplot)
sns.stripplot(
    x='feature', 
    y='value',
    hue="grupo",
    palette=palette,
    data=df, 
    jitter=True, 
    zorder=2,
    ax=ax
)

plt.title('Feature Importance Auditory Task')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.xticks(rotation=90)


fig.savefig('/home/nicobruno/flor/Tesis/graficos/feature_importance_auditory.png')
```
